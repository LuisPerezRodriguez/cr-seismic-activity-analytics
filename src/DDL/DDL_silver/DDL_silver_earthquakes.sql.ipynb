```python
%sql
-- Create the Silver cleaned and enriched seismic table
DROP TABLE IF EXISTS cr_seismic_analysis.silver.seismic_clean;

CREATE TABLE IF NOT EXISTS cr_seismic_analysis.silver.seismic_clean (
    seismic_event_id BIGINT GENERATED ALWAYS AS IDENTITY (START WITH 1 INCREMENT BY 1) COMMENT 'Unique earthquake identifier',
    event_id STRING COMMENT 'Unique earthquake identifier from USGS',
    magnitude DECIMAL(3,1) COMMENT 'Cleaned seismic magnitude value (e.g., 4.2)',
    magnitude_type STRING COMMENT 'Type of seismic scale used (e.g., MW, MB, ML)',
    magnitude_class STRING COMMENT 'Business logic classification: Minor, Light, Moderate, Strong, Major, Great',
    place STRING COMMENT 'Cleaned textual description of the geographic location',
    event_timestamp TIMESTAMP COMMENT 'Properly parsed event timestamp in UTC timezone',
    longitude DOUBLE COMMENT 'Geographic coordinate longitude (kept DOUBLE for GIS compatibility)',
    latitude DOUBLE COMMENT 'Geographic coordinate latitude (kept DOUBLE for GIS compatibility)',
    depth DECIMAL(5,1) COMMENT 'Depth of the seismic event in kilometers',
    source_file STRING COMMENT 'Audit metadata: Name of the raw source JSON file',
    ingestion_timestamp TIMESTAMP COMMENT 'Audit metadata: System timestamp when written to Silver'
)
USING DELTA
TBLPROPERTIES (
    'delta.enableChangeDataFeed' = 'true',
    'delta.autoOptimize.optimizeWrite' = 'true',
    'delta.autoOptimize.autoCompact' = 'true',
    'delta.feature.allowColumnDefaults' = 'supported'
)
COMMENT 'Silver table containing cleaned, strongly typed, and geofiltered seismic events for Costa Rica';
```

```python
%sql
DESCRIBE TABLE EXTENDED cr_seismic_analysis.silver.seismic_clean;
```
