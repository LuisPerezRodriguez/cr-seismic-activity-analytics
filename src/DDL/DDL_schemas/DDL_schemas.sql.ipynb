# Setup: Catalog Creation and Schema Definition

This notebook initializes the data platform environment by creating a dedicated catalog for the project and defining the Medallion schemas (Landing, Bronze, Silver, and Gold) to ensure structured data governance and organization.

```python
%sql
--1. Create the main catalog for the project
CREATE CATALOG IF NOT EXISTS cr_seismic_analysis;

--2. Set the current execution context to the new catalog
USE CATALOG cr_seismic_analysis;

--3. Create schemas for Medallion architecture
CREATE SCHEMA IF NOT EXISTS cr_seismic_analysis.landing
COMMENT 'Schema for RAW API ingestion';

CREATE SCHEMA IF NOT EXISTS cr_seismic_analysis.bronze
COMMENT 'Schema for data with generic types';

CREATE SCHEMA IF NOT EXISTS cr_seismic_analysis.silver
COMMENT 'Schema for cleaned and deduplicated data';

CREATE SCHEMA IF NOT EXISTS cr_seismic_analysis.gold
COMMENT 'Schema for dimensional modeling';
```
